```python
import pandas as pd
```

```python
df=pd.read_csv("worldcities.csv")
```

```python
df.columns
```

```python
harjoitus_a = df.sort_values(by=['population'],ascending=False).head(10)
harjoitus_a
```

```python
harjoitus_a["city"]
```

```python
harjoitus_b = df[(500000 < df['population']) & ( df['population']< 1000000)]
harjoitus_b
```

```python
harjoitus_b ["city"].unique()
```

## TOINEN TAPA

```python
for indeksi, rivi in df.iterrows():
    if rivi["population"] > 500000 and rivi["population"] < 1000000:
        print(rivi["city"], rivi["population"])
```

```python
# Harjoitus 2. SciKit-Learn ja hinnat.
# Asenna SciKit-Learn ja kirjoita koodia, joka käyttää lineaarista regressiota selvittämään hintojen kehityksen.
# Käytä opetusaineistona Hinnat.csv-tiedostoa.
# Mitä malli ennustaa hinnoiksi vuosille 2020-2023?
# Vinkki: tarvitset toimintoja "LinearRegression", "fit" ja "predict".
```

```python
hinnat = pd.read_csv("Hinnat.csv")
hinnat
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = hinnat[["vuosi"]]
y = hinnat["hinta"]
```

```python
X
```

```python
reg_hinnat = LinearRegression().fit(X, y)
```
